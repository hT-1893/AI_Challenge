```python
IMG_DIR = "download/train/"
imgs = os.listdir(IMG_DIR)
train = []
val = []
test = []

train_ratio = 0.8
val_ratio = 0.1

for i in range(1, 118):
    class_img = [img for img in imgs if int(img[:5]) == i]
    total_size = len(class_img)
    train_size = int(train_ratio * total_size)
    val_size = int(val_ratio * total_size)
    random.shuffle(class_img)
    train.extend(class_img[:train_size])
    val.extend(class_img[train_size:train_size + val_size])
    test.extend(class_img[train_size + val_size:])

IMG_TRAIN = 'data/train/'
for img in train:
    os.system('cp ' + IMG_DIR + img + ' ' + IMG_TRAIN)


len(train), len(val), len(test)
```
